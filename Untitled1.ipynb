```python
from osgeo import gdal
import rasterio
import leafmap
from samgeo import SamGeo
from samgeo.common import tms_to_geotiff
```

```python
m = leafmap.Map(center=[38.52053,-7.72861], zoom=15)
m.add_basemap("SATELLITE")
m
```

```python

if m.user_roi_bounds() is not None:
    bbox = m.user_roi_bounds()
else:
    bbox = [38.50674, -7.7697 , 38.53245, -7.7216]
```

```python
image = "satellite2.tif"
bbox
```

```python
tms_to_geotiff(output=image, bbox=bbox, zoom=20, source="Satellite", overwrite=True)
```

```python
m.layers[-1].visible = False  # turn off the basemap
m.add_raster(image, layer_name="Image")
m
```

```python
sam = SamGeo(
    model_type="vit_h",
    checkpoint="sam_vit_h_4b8939.pth",
    sam_kwargs=None,
)
```

```python


mask = "segment.tif"
sam.generate(
    image, mask, batch=True, foreground=True, erosion_kernel=(3, 3), mask_multiplier=255
)
```

```python


vector = "segment2.gpkg"
sam.tiff_to_gpkg(mask, vector, simplify_tolerance=None)
```

```python


sam = SamGeo(
    model_type="vit_h",
    sam_kwargs=None,
)
```

```python
sam.generate(image, batch=True, output="masks.tif", foreground=True, unique=True)
```
